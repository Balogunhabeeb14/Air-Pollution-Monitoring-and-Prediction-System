## Extracting Weather Data

```python
new_weather_data= pd.DataFrame()
def dt_datetime(dt):
    return datetime.datetime.strptime(dt, "%d/%m/%y %H:%M:%S")
def datetime_epoch(dt):
    return dt.timestamp()
for i,j in enumerate(postcodes):
    start_datetime = "01/05/22 00:00:00"
    end_datetime = "31/05/22 23:00:00"
    week_number = 1
    print(f'Extracting weather Data for: {j}')
    while(datetime_epoch(dt_datetime(start_datetime)) < datetime_epoch(dt_datetime(end_datetime))):
        #print('Week ',week_number)
        #print('Start Datetime: ',start_datetime)
        week_start_epoch = datetime_epoch(dt_datetime(start_datetime))
        week_end = dt_datetime(start_datetime)+ datetime.timedelta(hours=168)
        week_end_epoch = datetime_epoch(week_end)
        if week_end_epoch > datetime_epoch(dt_datetime(end_datetime)):
            week_end_epoch = datetime_epoch(dt_datetime(end_datetime))
        nomi = pgeocode.Nominatim('gb')
        da=nomi.query_postal_code(j)
        api_key = "fdcb3e66a02f57febe17dac02c44ed50"
        la = float(da.latitude)
        lo = float(da.longitude)
        if j == 'S5 7AA':
            la = 53.40546
            lo = -1.45549
        url ='http://history.openweathermap.org/data/2.5/history/city?lat=%s&lon=%s&start=%s&end=%s&units=metric&type=hour&appid=fdcb3e66a02f57febe17dac02c44ed50'%(la,lo,str(int(week_start_epoch)),str(int(week_end_epoch)))
        response = requests.get(url)
        pp = pd.DataFrame.from_dict(json.loads(response.text)['list'])
        dt = pp['dt'].apply(pd.Series)
        main = pp['main'].apply(pd.Series)
        wind = pp['wind'].apply(pd.Series)
        clouds = pp['clouds'].apply(pd.Series)
        weather = pp['weather'].apply(pd.Series)
        e = weather[0].apply(pd.Series)
        w1 = pd.concat([dt,main,wind,clouds,e],axis=1)
        w1.rename(columns = {0:'dt'}, inplace = True)
        w1['postcode']= postcodes[i].replace(' ','')
        new_weather_data= new_weather_data.append(w1)
        next_week_start = dt_datetime(start_datetime)+ datetime.timedelta(hours=169) 
        start_datetime = next_week_start.strftime("%d/%m/%y %H:%M:%S") 
        week_number += 1
    
new_weather_data
```

## Processing GIS Data

```python
def dataProcess(pc_topo_bh,road_nodes):
    columns = ['OBJECTID','theme','calculatedareavalue','abshmin', 'absh2','abshmax', 'relh2', 'relhmax', 'POSTCODE']
    pc_topo_bh = pc_topo_bh[columns]
    df1 = road_nodes.drop_duplicates(subset='identifier')

    junc_count= df1.formOfNode[df1.formOfNode == 'junction'].count()
    pseudo_count = df1.formOfNode[df1.formOfNode == 'pseudo node'].count()
    road_end_count = df1.formOfNode[df1.formOfNode == 'road end'].count()
    roundabout_count = df1.formOfNode[df1.formOfNode == 'roundabout'].count()

    df2 = pc_topo_bh.groupby(['OBJECTID','theme','POSTCODE'])[['calculatedareavalue','abshmin', 'absh2','abshmax', 'relh2', 'relhmax']].mean().reset_index()
    df2['average_height'] = df2[['abshmin', 'absh2','abshmax', 'relh2', 'relhmax']].mean(axis=1)
    df2.loc[((df2['theme'] == 'Buildings')&(df2['average_height'] <= 7)), 'building_classification'] = 'lr'
    df2.loc[((df2['theme'] == 'Buildings')&(df2['average_height'] > 7)& (df2['average_height'] <= 18)), 'building_classification'] = 'mr'
    df2.loc[((df2['theme'] == 'Buildings')&(df2['average_height'] > 18)& (df2['average_height'] <= 30)), 'building_classification'] = 'hr'
    df2.loc[((df2['theme'] == 'Buildings')&(df2['average_height'] > 30)), 'building_classification'] = 'vhr'

    building_count= df2.theme[df2.theme == 'Buildings'].count()
    building_area = df2.calculatedareavalue[df2.theme == 'Buildings'].sum()
    green_area= df2.calculatedareavalue[df2.theme == 'Land'].sum()
    green_count= df2.theme[df2.theme == 'Land'].count()
    roads_tracks_area= df2.calculatedareavalue[df2.theme == 'Roads Tracks And Paths'].sum()
    roads_track_count= df2.theme[df2.theme == 'Roads Tracks And Paths'].count()
    water_area= df2.calculatedareavalue[df2.theme == 'water'].sum()
    water_count= df2.theme[df2.theme == 'water'].count()

    rail_area= df2.calculatedareavalue[df2.theme == 'rail'].sum()
    rail_count= df2.theme[df2.theme == 'rail'].count()

    no_of_lr = df2.theme[df2.building_classification == 'lr'].count()
    no_of_mr = df2.theme[df2.building_classification == 'mr'].count()
    no_of_hr = df2.theme[df2.building_classification == 'hr'].count()
    no_of_vhr = df2.theme[df2.building_classification == 'vhr'].count()

    values = {'postcode':df2.POSTCODE[1].replace(' ',''),'building_count':building_count,'no_of_lr':no_of_lr,'no_of_mr':no_of_mr,
              'no_of_hr':no_of_hr, 'no_of_vhr':no_of_vhr,'building_area':building_area,'green_area':green_area,
              'green_count':green_count,'roads_tracks_area':roads_tracks_area,'roads_track_count':roads_track_count,
              'junc_count':junc_count,'pseudo_count':pseudo_count,'road_end_count':road_end_count,'roundabout_count':roundabout_count,
             'water_area':water_area,'water_count':water_count,'rail_area':rail_area,'rail_count':rail_count}
    temp = pd.DataFrame(values, index=[0])
    
    return temp
    
    
```
